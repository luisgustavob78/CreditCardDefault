```python
import csv
import json

csv_file_path = '../batches/batch_0.csv'
json_file_path = '../batches/batch.json'

data = []

with open(csv_file_path, 'r') as csv_file:
    csv_reader = csv.reader(csv_file)
    header_skipped = False

    for row in csv_reader:
        if not header_skipped:
            header_skipped = True
            continue
        
        data.append(row)

# Wrap the data in a dictionary with the "batches" key
json_data = json.dumps({"batches": data})

# Save the JSON data to a file
with open(json_file_path, 'w') as json_file:
    json_file.write(json_data)

print("CSV data has been converted and saved as JSON with the desired structure.")
```
